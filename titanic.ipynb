```python


import pandas as pd           # Data manipulation
import numpy as np            # Numerical operations
import matplotlib.pyplot as plt  # Visualization
import seaborn as sns         # Advanced visualization
```

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
titanic.head()
```

```python
# Display basic information and statistics about the dataset
print("Dataset Info:")
titanic.info()
print("\nSummary Statistics:")
display(titanic.describe(include='all'))
print("\nMissing Values:")
print(titanic.isnull().sum())
```

```python
# Quick overview of all columns: name, data type, number of records, missing values
overview = pd.DataFrame({
    'Data Type': titanic.dtypes,
    'Non-Null Count': titanic.count(),
    'Missing Values': titanic.isnull().sum(),
    'Total Records': len(titanic)
})
display(overview)
```

```python
# Display the number of missing values per column
missing_values = titanic.isnull().sum()
print(missing_values)
```

```python
# Univariate EDA
print("Univariate Analysis:\n")
for col in titanic.columns:
    print(f"--- {col} ---")
    if titanic[col].dtype in ['int64', 'float64']:
        print(titanic[col].describe())
        plt.figure(figsize=(6, 2))
        sns.histplot(titanic[col].dropna(), kde=True)
        plt.title(f'Distribution of {col}')
        plt.show()
    else:
        print(titanic[col].value_counts(dropna=False))
        plt.figure(figsize=(6, 2))
        sns.countplot(y=titanic[col])
        plt.title(f'Countplot of {col}')
        plt.show()
    print("\n")

# Bivariate EDA
print("Bivariate Analysis:\n")
# Numerical vs Target
num_cols = titanic.select_dtypes(include=['int64', 'float64']).columns.drop('Survived')
for col in num_cols:
    plt.figure(figsize=(6, 3))
    sns.boxplot(x='Survived', y=col, data=titanic)
    plt.title(f'{col} vs Survived')
    plt.show()

# Categorical vs Target
cat_cols = titanic.select_dtypes(include='object').columns
for col in cat_cols:
    plt.figure(figsize=(6, 3))
    sns.countplot(x=col, hue='Survived', data=titanic)
    plt.title(f'Survived by {col}')
    plt.xticks(rotation=45)
    plt.show()

# Multivariate EDA
print("Multivariate Analysis:\n")
plt.figure(figsize=(10, 8))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Pairplot for selected features
selected = ['Survived', 'Pclass', 'Age', 'Fare', 'SibSp', 'Parch']
sns.pairplot(titanic[selected], hue='Survived')
plt.show()

# Insights
print("Insights:\n")
for col in titanic.columns:
    print(f"Column: {col}")
    if titanic[col].dtype in ['int64', 'float64']:
        print(f"  Mean: {titanic[col].mean():.2f}, Median: {titanic[col].median():.2f}, Std: {titanic[col].std():.2f}")
        print(f"  Missing: {titanic[col].isnull().sum()} ({titanic[col].isnull().mean()*100:.1f}%)")
    else:
        top = titanic[col].value_counts(dropna=False).idxmax()
        freq = titanic[col].value_counts(dropna=False).max()
        print(f"  Most frequent: {top} ({freq} times)")
        print(f"  Missing: {titanic[col].isnull().sum()} ({titanic[col].isnull().mean()*100:.1f}%)")
    print()
```

```python
numerical_cols = titanic.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = titanic.select_dtypes(include='object').columns.tolist()
print("Numerical columns:", numerical_cols)
print("Categorical columns:", categorical_cols)
```

```python
# Correlation heatmap for numerical columns
plt.figure(figsize=(8, 6))
corr_matrix = titanic[numerical_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Columns')
plt.show()
```

```python
# a. Handle missing values
# For 'Age', fill missing values with the median
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
# For 'Fare', fill missing values with the median (if any)
if titanic['Fare'].isnull().any():
    titanic['Fare'].fillna(titanic['Fare'].median(), inplace=True)
# For 'Embarked', fill missing values with the mode (only if column exists)
if 'Embarked' in titanic.columns:
    titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace=True)

# b. Convert categorical variables to numeric
# Encode 'Sex' as 0/1
titanic['Sex'] = titanic['Sex'].map({'male': 0, 'female': 1})
# One-hot encode 'Embarked' (only if column exists)
if 'Embarked' in titanic.columns:
    embarked_dummies = pd.get_dummies(titanic['Embarked'], prefix='Embarked', drop_first=True)
    titanic = pd.concat([titanic, embarked_dummies], axis=1)

# c. Drop features not useful for the model (do this last)
# Drop features not useful for the model
# Drop features not useful for the model, but only if they exist in the DataFrame
cols_to_drop = ['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked']
titanic.drop([col for col in cols_to_drop if col in titanic.columns], axis=1, inplace=True)
```
